### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
import matplotlib.pyplot as plt

ingresosTotalTienda =  tienda['Precio'].sum()
ingresosTotalTienda2 =  tienda2['Precio'].sum()
ingresosTotalTienda3 =  tienda3['Precio'].sum()
ingresosTotalTienda4 =  tienda4['Precio'].sum()

print(f"Ingreso total de la primera tienda: {ingresosTotalTienda}")
print(f"Ingreso total de la primera tienda 2: {ingresosTotalTienda2}")
print(f"Ingreso total de la primera tienda 3: {ingresosTotalTienda3}")
print(f"Ingreso total de la primera tienda 4: {ingresosTotalTienda4}")


```

# 2. Ventas por categoría

```python
productos_por_categoria = tienda.groupby('Categoría del Producto')['Precio'].count()
productos_por_categoria_Ordenada = productos_por_categoria.sort_values(ascending= False)

productos_por_categoria2 = tienda2.groupby('Categoría del Producto')['Precio'].count()
productos_por_categoria_Ordenada2 = productos_por_categoria2.sort_values(ascending= False)

productos_por_categoria3 = tienda3.groupby('Categoría del Producto')['Precio'].count()
productos_por_categoria_Ordenada3 = productos_por_categoria3.sort_values(ascending= False)

productos_por_categoria4 = tienda4.groupby('Categoría del Producto')['Precio'].count()
productos_por_categoria_Ordenada4 = productos_por_categoria4.sort_values(ascending= False)

print('tienda 1')
print(productos_por_categoria_Ordenada)
print('')
print('tienda 2')
print(productos_por_categoria_Ordenada2)
print('')
print('tienda 3')
print(productos_por_categoria_Ordenada3)
print('')
print('tienda 4')
print(productos_por_categoria_Ordenada4)
```

# 3. Calificación promedio de la tienda


```python
clasificacionPromedio_tienda = tienda['Calificación'].mean()
clasificacionPromedio_tienda2 = tienda2['Calificación'].mean()
clasificacionPromedio_tienda3 = tienda3['Calificación'].mean()
clasificacionPromedio_tienda4 = tienda4['Calificación'].mean()

print(f"tienda: {clasificacionPromedio_tienda}")
print(f"tienda2: {clasificacionPromedio_tienda2}")
print(f"tienda3: {clasificacionPromedio_tienda3}")
print(f"tienda4: {clasificacionPromedio_tienda4}")
```

# 4. Productos más y menos vendidos

```python
producto_nombres = tienda.groupby('Producto')['Precio'].count()
productos_por_nombres_Ordenada = producto_nombres.sort_values(ascending= False)
productosMasVendidos = productos_por_nombres_Ordenada.head()
productosMenosVendidos = productos_por_nombres_Ordenada.tail()

producto_nombres2 = tienda2.groupby('Producto')['Precio'].count()
productos_por_nombres_Ordenada2 = producto_nombres2.sort_values(ascending= False)
productosMasVendidos2 = productos_por_nombres_Ordenada2.head()
productosMenosVendidos2 = productos_por_nombres_Ordenada2.tail()

producto_nombres3 = tienda3.groupby('Producto')['Precio'].count()
productos_por_nombres_Ordenada3 = producto_nombres3.sort_values(ascending= False)
productosMasVendidos3 = productos_por_nombres_Ordenada3.head()
productosMenosVendidos3 = productos_por_nombres_Ordenada3.tail()

producto_nombres4 = tienda4.groupby('Producto')['Precio'].count()
productos_por_nombres_Ordenada4 = producto_nombres4.sort_values(ascending= False)
productosMasVendidos4 = productos_por_nombres_Ordenada4.head()
productosMenosVendidos4 = productos_por_nombres_Ordenada4.tail()

print('tienda 1')
print(productosMasVendidos)
print('')
print(productosMenosVendidos)
print('')

print('tienda 2')
print(productosMasVendidos2)
print('')
print(productosMenosVendidos2)
print('')

print('tienda 3')
print(productosMasVendidos3)
print('')
print(productosMenosVendidos3)
print('')

print('tienda 4')
print(productosMasVendidos4)
print('')
print(productosMenosVendidos4)
print('')
```

# 5. Envío promedio por tienda

```python
constoEnvioPromedio = tienda['Costo de envío'].mean()
constoEnvioPromedio2 = tienda2['Costo de envío'].mean()
constoEnvioPromedio3 = tienda3['Costo de envío'].mean()
constoEnvioPromedio4 = tienda4['Costo de envío'].mean()


print(f"tienda: {constoEnvioPromedio}")
print(f"tienda2: {constoEnvioPromedio2}")
print(f"tienda3: {constoEnvioPromedio3}")
print(f"tienda4: {constoEnvioPromedio4}")
```

```python
# Nombres de las tiendas
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Valores de costos de envío promedio
costos_envio = [constoEnvioPromedio, constoEnvioPromedio2, constoEnvioPromedio3, constoEnvioPromedio4]

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(tiendas, costos_envio, color='skyblue')

# Agregar título y etiquetas
plt.title('Costo de Envío Promedio por Tienda', fontsize=16)
plt.xlabel('Tiendas', fontsize=12)
plt.ylabel('Costo de Envío Promedio', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Mostrar los valores en cada barra
for i, valor in enumerate(costos_envio):
    plt.text(i, valor + 0.5, f'{valor:.2f}', ha='center', fontsize=10)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
# Función para generar un gráfico circular de porcentaje por categoría
def graficar_pastel(categorias, tienda_nombre):
    total = categorias.sum()
    porcentajes = (categorias / total) * 100

    # Crear el gráfico circular
    plt.figure(figsize=(8, 8))
    plt.pie(porcentajes, labels=categorias.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
    plt.title(f'Porcentaje de Productos Vendidos por Categoría - {tienda_nombre}', fontsize=14)
    plt.tight_layout()

    # Mostrar el gráfico
    plt.show()

# Generar gráficos para cada tienda
print('Generando gráficos circulares...')
graficar_pastel(productos_por_categoria_Ordenada, 'Tienda 1')
graficar_pastel(productos_por_categoria_Ordenada2, 'Tienda 2')
graficar_pastel(productos_por_categoria_Ordenada3, 'Tienda 3')
graficar_pastel(productos_por_categoria_Ordenada4, 'Tienda 4')
```
